```python
import nibabel as nib
import matplotlib.pyplot as plt
#108?
# Load the NIfTI files
nifti_0 = nib.load('D:/ismi_data/NLST/imagesVal/NLST_0104_0000.nii.gz')
nifti_1 = nib.load('D:/ismi_data/NLST/imagesVal/NLST_0104_0001.nii.gz')
nifti_2 = nib.load('D:/ismi_data/base_results/dist_0104_0104_image.nii.gz')
nifti_3 = nib.load('D:/ismi_data/results/disp_0104_0104_image.nii.gz')

# Get the image data as 3D numpy arrays
image_data_0 = nifti_0.get_fdata()
image_data_1 = nifti_1.get_fdata()
image_data_2 = nifti_2.get_fdata()
image_data_3 = nifti_3.get_fdata()

# Set the figure size and create a 3x3 subplot grid
fig, axes = plt.subplots(3, 4, figsize=(12, 12))

# Plot the images in the subplot grid
axes[0, 0].imshow(image_data_0[100, :, :], cmap='gray')
axes[0, 0].set_title('Sagittal - Fixed image')
axes[0, 1].imshow(image_data_1[100, :, :], cmap='gray')
axes[0, 1].set_title('Sagittal - Moving image')
axes[0, 2].imshow(image_data_2[100, :, :], cmap='gray')
axes[0, 2].set_title('Sagittal - Base model aligned image')
axes[0, 3].imshow(image_data_3[100, :, :], cmap='gray')
axes[0, 3].set_title('Sagittal - Improved model aligned image')

axes[1, 0].imshow(image_data_0[:, 100, :], cmap='gray')
axes[1, 1].imshow(image_data_1[:, 100, :], cmap='gray')
axes[1, 2].imshow(image_data_2[:, 100, :], cmap='gray')
axes[1, 3].imshow(image_data_3[:, 100, :], cmap='gray')

axes[2, 0].imshow(image_data_0[:, :, 100], cmap='gray')
axes[2, 1].imshow(image_data_1[:, :, 100], cmap='gray')
axes[2, 2].imshow(image_data_2[:, :, 100], cmap='gray')
axes[2, 3].imshow(image_data_3[:, :, 100], cmap='gray')


# Remove axis ticks and labels from all subplots
for ax in axes.flatten():
    ax.axis('off')

# Set titles for each row
axes[0, 0].set_title('Sagittal - Fixed')
axes[0, 1].set_title('Sagittal - Moving')
axes[0, 2].set_title('Sagittal - Base model aligned')
axes[0, 3].set_title('Sagittal - Improved model aligned')
axes[1, 0].set_title('Coronal - Fixed ')
axes[1, 1].set_title('Coronal - Moving ')
axes[1, 2].set_title('Coronal - Base model aligned ')
axes[1, 3].set_title('Coronal - Improved model aligned ')
axes[2, 0].set_title('Transverse - Fixed ')
axes[2, 1].set_title('Transverse - Moving ')
axes[2, 2].set_title('Transverse - Base model aligned ')
axes[2, 3].set_title('Transverse - Improved model aligned ')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the subplot grid
plt.show()
```
